```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, expon, binom, poisson

```

```python
# Uniform Distributions
# Equal chance of the different values, from -10 to 10, of ocurring. 
# This Histogram shows the number of ocurrences for each buckect (number/value)
# Flat line on a Probability Density Function
values = np.random.uniform(-10.0, 10.0, 100000)
plt.hist(values, 50)
plt.show()
```

```python
# Normal / Gausian Distributions

x = np.arange(-3, 3, 0.001)
plt.plot(x, norm.pdf(x))
plt.show()

Y = norm.pdf(x)
plt.plot(x, Y)
plt.show()
```

```python
mu = 5
sigma = 2

data = np.random.normal(mu, sigma, 10000)

plt.hist(data, 50)
plt.show()
```

```python
# Exponential PDF (Probability Density Function)

x = np.arange(0, 10, 0.001)
plt.plot(x, expon.pdf(x))
plt.show()
```

```python
n, p = 10, 0.5
x = np.arange(0, 10, 0.001)
plt.plot(x, binom.pmf(x, n, p))
plt.show()
```

```python
mu = 500
x = np.arange(400, 600, 0.5)
plt.plot(x, poisson.pmf(x, mu))
plt.show()
```
