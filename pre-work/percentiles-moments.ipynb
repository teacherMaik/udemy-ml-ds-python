```python
import numpy as np
import scipy.stats as sp
import matplotlib.pyplot as plt
```

```python
vals = np.random.normal(0, 0.5, 10000)

plt.hist(vals, 50)
plt.show()

print(np.mean(vals))
print(np.percentile(vals, 50))
print(np.percentile(vals, 90))
print(np.percentile(vals, 99))
```

```python
# Moments

# First moment is the mean
moment_1 = np.mean(vals)
print(moment_1)

# Second moment is the Variance
moment_2 = np.var(vals)
print(moment_2)

# Moment 3 is the skew
moment_3 = sp.skew(vals)
print(moment_3)

# Moment 4 is the "Kurtosis"
moment_4 = sp.kurtosis(vals)
print(moment_4)
```
